```python
import cv2 as cv
import Mask2PolyConverter as Mask2Poly
import matplotlib.pyplot as plt
```

```python

#LOAD original
im = cv.imread('Sample_3.png') #Remplacer par

print("[Original Image]")
plt.imshow(im)
assert im is not None, "file could not be read, check with os.path.exists()"
```

```python
#PREPARE image to find contours
#imgray = cv.cvtColor(im, cv.COLOR_RGB2GRAY)
#cv.imshow("gray",imgray)


# FIND edges


#CONVERT edges to polygons
polys = Mask2Poly.mask_to_polygons(im)
print("Nb de polygones "+str(len(polys.geoms)))
```

Maintenant qu'on a un tableau Multipolygon, on peut le convertir Ã  nouveau en masque avec Polygons_To_Mask

```python

#CONVERT the polygons back to a mask image to validate that all went well
mask = Mask2Poly.Polygons_To_Mask(polys.geoms, im.shape[:2]) #attention au .geom si nÃ©cÃ©ssaire

#cv.imshow("final mask",mask)
print("Final Image")
plt.imshow(mask)
cv.imwrite("Result.png",mask)
```
